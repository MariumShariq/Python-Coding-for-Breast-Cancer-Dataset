```python
#Data Visualization
# Day 3 Challenge
# plt.bar(x axis, y axis)

condition_breakdown = df['Patient Status'].value_counts()
bar_xaxis = condition_breakdown.index

bar_yaxis = condition_breakdown.values

plt.bar(bar_xaxis, bar_yaxis)
plt.title("Patients Condition")

plt.xlabel("Status")
plt.ylabel("Count of Patients")
plt.show()
```

```python
# Pie Chart
plt.pie(bar_yaxis, labels=bar_xaxis,autopct='%1.1f%%')
plt.title("Patients Condition")
plt.show()
```

```python
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count of Patients')
plt.show()
```

```python
pd.crosstab(df['Surgery_type'], df['Patient Status']).plot(kind='bar', stacked=True)
plt.title('Surgery Type by Patient Status')
plt.xlabel('Surgery Type')
plt.ylabel('Count')
plt.show()
```

```python

#Day 4 Challenge

# Select relevant numerical columns
numeric_columns = ['Age', 'Protein1', 'Protein2', 'Protein3', 'Protein4']

# Descriptive statistics for the numerical columns
descriptive_stats = df[numeric_columns].describe()

modes = df[numeric_columns].mode().iloc[0]

# Adding mode to descriptive statistics output
descriptive_stats.loc['mode'] = modes

# Display the descriptive statistics
print(descriptive_stats)
```

```python
# Histogram for Age Distribution
fig_age = px.histogram(df, x='Age', nbins=10, title='Age Distribution of Patients')
fig_age.update_layout(xaxis_title='Age', yaxis_title='Count of Patients')
fig_age.show()
```
